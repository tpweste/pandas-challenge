```python
file = "02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"

purchase_data = pd.read_csv(file)

purchase_data.head()
```

```python
total_players = purchase_data["SN"].value_counts()
total_players
```

```python
total_players_df = pd.DataFrame({
    "Total Players":[len(total_players)]
})

total_players_df.head()
```

```python
unique_items = purchase_data["Item ID"].value_counts()
unique_items
```

```python
average_price = purchase_data["Price"].mean()
average_price
```

```python
number_of_purchases = purchase_data["Purchase ID"].value_counts()
number_of_purchases
```

```python
total_revenue = purchase_data["Price"].sum()
total_revenue
```

```python
purchase_analysis_df = pd.DataFrame({
    "Number of Unique Items":[len(unique_items)],
    "Average Price":[average_price],
    "Number of Purchases":[len(number_of_purchases)],
    "Total Revenue":[total_revenue]
})
purchase_analysis_df['Average Price'] = pd.Series([round(val, 2) for val in purchase_analysis_df['Average Price']], index = purchase_analysis_df.index)
purchase_analysis_df['Total Revenue'] = purchase_analysis_df['Total Revenue'].apply(lambda x: "${:.2f}".format((x/1)))
purchase_analysis_df.head()
```

```python
gender_total = gender_demo_df["Gender"].value_counts()
gender_total
```
